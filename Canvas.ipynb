#### 2. Интервью с заказчиком и понимание задачи
Задача финансового отдела департамента HR – эффективное распределение премий.


HR менеджеры распределены по должностям сотрудников и каждый отвечает за свою должность.


Основная метрика оценки сотрудников: performance rating. Он ставится от 1 до 4: 1 – худший, 4 – лучший.
Премии сотрудникам в компании считаются в зависимости от рейтинга и количества завершенных проектов по следующему принципу:
* Рейтинг ниже среднего, менее 15 проектов: 5-8%
* Рейтинг ниже среднего, более 15 проектов: 8-10%
* Рейтинг выше среднего, менее 15 проектов: 10-15%
* Рейтинг выше среднего, более 15 проектов: 15-20%

HR менеджер уже сам выбирает процент премии для каждого сотрудника из представленного диапазона.


Дашборд необходим, чтобы удобно считать премии по каждому сотруднику, а также чтобы в целом понимать и оценивать перформанс сотрудников и рассчитывать бюджет на следующий год.

#### 3. Портрет пользователя
Кто будет пользоваться дашбородом? Какие их роли и задачи? Кто будет входить в тестовую группу?

**Руководитель финансового отдела.** Его задачи:
* Оценить в целом перформанс сотрудников. Для этого нужно:
* Общее распределение рейтинга
* Количество людей с рейтингом больше и меньше среднего
* Уровень зарплаты
* Распределение премий
* Овервью по количеству закрытых проектов в разрезе департаментов и должностей
* Спрогнозировать бюджет премий на следующий год. Для этого нужно:
* Посчитать общую сумму премии за текущий период по максимально возможной процентной ставке.
* Калькулятор, позволяющий посчитать бюджет премий при изменении границ процентов премий.

**HR менеджер**. Его задача – посчитать премии каждому сотруднику должности, за которую ответственен HR менеджер. Для этого ему нужна следующая информация: самооценка сотрудником удовлетворенности работой, стаж работы и уровень зарплаты. Еще HR менеджеру нужно знать: рейтинг каждого сотрудника, больше он или меньше среднего по должности, количество закрытых проектов, уровень удовлетворенности работой и зарплату. 

В тестовую группу будут входить HR менеджеры и руководитель.

#### 5. Вопросы и бизнес-решения
* Какова максимальная сумма премий за прошедший период по максимальной процентной ставке?
* Если поменять границы премий, то как изменится бюджет?
* Как распределились сотрудники по рейтингу?
* Какое количество сотрудников в департаменте R&D получили рейтинг 3?
* Сколько сотрудников получило оценку выше среднего?
* Какой уровень зарплаты у сотрудников с оценкой ниже среднего?
* Какие премии получаю сотрудники с зарплатой выше среднего?
* Какое количество закрытых проектов у Sales Executive?
* Сотрудники какой должности имеют больший вклад в общее количество закрытых проектов?
* Департамент R&D закрывают проектов больше или меньше среднего значения по компании?
* Какой диапазон премии у каждого сотрудника конкретной должности?

#### 6. Данные
Есть таблица в Excel с данными по каждому сотруднику с проставленным по каждому сотруднику рейтингу, количеству проектов и тп.
#### 7. Основные блоки отчета и типы графиков, подбор форматов
Для руководителя:

* Блок «Бюджет» с калькулятором изменения процентов премий: фактоид + калькулятор
* Блок «Перформанс сотрудников». Барчарт, показывающий количество сотрудников на каждое значение рейтинга в разрезе департаментов. Барчарты, показывающие количество закрытых проектов и среднее количество проектов сотрудников в разрезе должностей и департаментов.
* Блок «Распределение премий». Таблицы с 2-мя рядами: рейтинг выше среднего и ниже среднего. В каждом ряду: количество сотрудников, средняя зарплата, гистограмма распределения зарплат, средняя премия и график (боксплот) распределения премий, на котором можно выбрать каждого сотрудника и посмотреть его IDи премию.

Для HR менеджеров:

* Блок «Детальные данные по сотрудникам» с выбором должности. Это таблица со всеми необходимыми данными по каждому сотруднику (в том числе с расчетным диапазоном процента премий) для расчета премии.
